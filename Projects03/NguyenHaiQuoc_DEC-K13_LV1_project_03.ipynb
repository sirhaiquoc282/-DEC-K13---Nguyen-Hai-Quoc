```python
import pandas as pd
```

```python
movies_df = pd.read_csv("tmdb-movies.csv")
```

### 1. Sắp xếp các bộ phim theo ngày phát hành giảm dần rồi lưu ra một file mới

```python
def datetime_format(time: str):
    datetime = time.split('/')
    year = int(datetime[2])
    if year >= 25 and year <= 99:
        year+=1900
    else:
        year+=2000
    return f"{datetime[0]}/{datetime[1]}/{year}"
        
```

```python
movies_df["release_date"] = pd.to_datetime(
    movies_df["release_date"].apply(datetime_format),
    format="%m/%d/%Y"
)
```

```python
df1 =  movies_df.sort_values(by="release_date", ascending=False)
df1.to_csv('question1.csv', index=False)
```

### 2. Lọc ra các bộ phim có đánh giá trung bình trên 7.5 rồi lưu ra một file mới

### 6. Đạo diễn nào có nhiều bộ phim nhất và diễn viên nào đóng nhiều phim nhất

```python
directors = movies_df['director'].str.split("|").explode()
```

```python
directors.head()
```

```python
top_director = directors.mode()[0]
print(f"Top director: {top_director}")
```

```python
actors = movies_df['cast'].str.split("|").explode()
```

```python
actors.head()
```

```python
top_actor = actors.mode()[0]
print(f"Top actor: {top_actor}")
```
